```python
colsel_merged.dropna(inplace=True)
```

# Cleaning dataset: 

```python
colsel_merged['body'] = colsel_merged['body'].str.replace(r'(\(Il Sole 24 Ore Radiocor Plus\).\-.*?)\-','')
colsel_merged['body'] = colsel_merged['body'].str.replace(r'© RIPRODUZIONE RISERVATA','')
colsel_merged['body'] = colsel_merged['body'].str.replace(r'(www\.)+[a-zA-Z0-9]+(\.com)','')
colsel_merged['body'] = colsel_merged['body'].str.replace('\n','')
colsel_merged['body'] = colsel_merged['body'].str.replace(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?','')
colsel_merged['body'] = colsel_merged['body'].apply(lambda x : str(x).lower())
print("SHAPE BEFORE REMOVING DUPLICATES: " + str(colsel_merged.shape))
colsel_merged.drop_duplicates(inplace=True)
print("SHAPE AFTER REMOVING DUPLICATES: " + str(colsel_merged.shape))
colsel_merged['body'] = colsel_merged['body'].apply(lambda x: BeautifulSoup(str(x), "lxml").get_text())
```

```python
colsel_merged.reset_index()
colsel_merged.drop('index', inplace=True)
colsel_merged
```

```python
colsel_merged[colsel_merged['identificativo']=='RR201611141655450']
tokenized_example = gensim.utils.simple_preprocess(colsel_merged.iloc[0,1])
```

```python
words = [w for w in tokenized_example if not w in stopwords.words("italian")]
words
```
